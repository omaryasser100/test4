# GIT
## GIT is a version control system
### 1-Download git
### 2-Login from vscode UI or from cmd using
### git config --global user.name "Your Name"
### git config --global user.email "youremail@gmail.com"

## Basic git commands
### 1-pwd: shows current working directory(folder)

### 2-ls: shows files in the current working directory(folder)

### 3-git init: initialize the folder for git 

### 4-git status:tracks the file status what is changed and what is the same

### 5-git add --all: moves a file or all files(that are changed) from the working directory to the
###            staging area

### 6-git commit -m "commit message": checkout the files to be uploaded with the upload message

### main: main branch we are working on
### origin: the repo link we working on
### git branch -M main :make the current branch main not master (github excpects main but git names it master)

### 7-git remote add origin repo_link:connect with the certain repo u are going to upload to
### check: git remote -v

### 8-git push -u origin main: upload all the files in the commit to the github repo
### hint -u only for first time to remember the repo

### 9-git pull origin main:
### if git refuses to push. it means unrelated histories are made in git and github
### solution git pull origin main --allow-unrelated-histories  then press esc and type :wq and enter
### this will merge unrelated histories and we are safe to go

### 10-git clone repo_url:downloades the contents of a repostory in ur PC as a new folder
### U MUST MOVE TO THE NEW REPO FOLDER WITH THE CMD TO UPDATE GIT USING CD COMMAND

### 11-git checkout -b branch_name: creates a new branch for u 
### check: git branch

# FINAL GITHUB FLOW:
## First time u set up a project (OPEN THE PROJECT FOLDER FROM THE VSCODE EXPLORER):
### 1-git init: To setup git to track the project
### 2-git status: To see the branch u are on and the files that is needed to be uploaded in the project folder
### 3-git branch -M main: Use to set the branch to main if it is set to master
### 4-git branch or git status:To check if the branch is named correctly
### 5-git remote add origin repo_url:set the repo we are uploading too
### 6-git add --all:add all the files to the staging area 
### 7-git commit -m "message":prepare the commit with the update message (FINAL STEP BEFORE PUSHING)
### 8-git push -u origin main:Upload the files to the repo
